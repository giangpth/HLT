```python
import numpy as np
import pandas as pd
import sys
import random

from tqdm import tqdm

import re
import string

import os

import shutil

import json
```

```python
f_train = '/content/drive/MyDrive/data/simplified-nq-train.jsonl'
f_test = '/content/drive/MyDrive/data/simplified-nq-test.jsonl'
num_train_samples = 307372
num_test_samples = 346
```

```python
def scanData(filename=f_train, debug=False):    
    num = 0
    num_pos = 0
    num_neg = 0

    list_pos = []
    list_neg = []
    
    with open(filename) as f:
        progress = tqdm(f, total=num_train_samples)
        for sam_count, line in enumerate(progress): 
            if sam_count >= num_train_samples:
                break

            num += 1    
            data = json.loads(line)
            
            example_id = str(data['example_id'])
            ans_id = data['annotations'][0]['long_answer']['candidate_index']
            if ans_id == -1:
                list_neg.append(example_id)
                num_neg += 1
            else:
                list_pos.append(example_id)
                num_pos += 1
    print("NUMBER SAMPLE = ", num)
    print("NUMBER POSITIVE SAMPLE = ", num_pos)
    print("NUMBER NEGATIVE SAMPLE = ", num_neg)   
    return list_pos, list_neg
```

```python
list_pos, list_neg = scanData()
```

```python
# get val set by 5% of positive and 5% of negative samples 
random.shuffle(list_pos)
random.shuffle(list_neg)

num_val_pos = len(list_pos)//100
num_val_neg = len(list_neg)//100

val_pos = list_pos[0:num_val_pos + 1]
val_neg = list_neg[0:num_val_neg + 1]

train_pos = list_pos[num_val_pos:len(list_pos)]
train_neg = list_neg[num_val_neg:len(list_neg)]

val = val_pos + val_neg
train = train_pos + train_neg 

print("Val set: ", len(val))
# random.shuffle(val)
print("Train set: ", len(train))
# random.shuffle(val)
```

```python
val_set = {'example_id': val}
val_df = pd.DataFrame(val_set)
val_df.to_csv('/content/drive/MyDrive/data/val_id.csv', index=False, columns=['example_id'])
```
